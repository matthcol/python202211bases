## Geo Pandas
https://geopandas.org/

```python
import geopandas as gpd
from shapely.geometry import Point
import osmnx as ox
```

```python
gdfToulouse = ox.geocode_to_gdf({'city': 'Toulouse'})
```

```python
def point_from_yx(yx):
    y,x = yx.split(',')
    x, y = float(x), float(y)
    return Point(x,y)
```

```python
coffees['geometry'] = coffees.geo_point.apply(point_from_yx)
coffeesGeo =  gpd.GeoDataFrame(coffees)
```

```python
ax = gdfToulouse.plot()
coffeesGeo.plot(color='red', ax=ax)
```
